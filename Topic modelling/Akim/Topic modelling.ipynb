```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   #Data visualisation libraries 

data = pd.read_csv('data.txt')
data
```

# Data Cleaning

## Merge columns starter_content and reply_content

```python
data_to_clean = {'Content': [str(starter) +' ' + str(reply) for (starter, reply) in zip(data.starter_content, data.reply_content)], 'Title': data.title}
df = pd.DataFrame(data_to_clean)
df
```

```python
df.Content[0]
```

```python
import re
import string

def clean_text_round1(text):
    '''Get rid of urls, lowercase everything, remove punctuation'''
    text = re.sub('http[s]?://\S+', '', text, flags=re.MULTILINE) # remove urls
    text = text.lower() # lowercase everything
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # remove punctuation
    text = re.sub('\w*\d\w*', '', text) # remove words containing numbers
    return text

round1 = lambda x: clean_text_round1(x) if type(x) != float else str(x)
```

```python
df.Content = df.Content.apply(round1)
df.Content[0]
```

```python
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    return text

round2 = lambda x: clean_text_round2(x) if type(x) != float else str(x)
```

```python
df.Content = df.Content.apply(round2)
df
```

# Document-Term Matrix

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
data_cv = cv.fit_transform(df.Content)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm
```

Oops, there are non-english characters!

```python
import nltk

words = set(nltk.corpus.words.words())

def clean_text_round3(text):
    '''Get rid of non-english characters.'''
    text = " ".join(w for w in nltk.wordpunct_tokenize(text) if w in words or not w.isalpha()) # remove non-english characters
    return text

round3 = lambda x: clean_text_round3(x)
```

```python
df.Content = df.Content.apply(round3)
```

```python
data_cv = cv.fit_transform(df.Content)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm
```

# Topic Modelling

```python
from nltk import pos_tag, word_tokenize

def nouns_adj(text):
    '''Given a string of text, tokenize the text and pull out only the nouns and adjectives.'''
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)]
    return ' '.join(nouns_adj)
```

```python
df.Content = df.Content.apply(nouns_adj)
```

```python
from gensim import matutils, models
import scipy.sparse

data_dtm.transpose().head()
```

```python
from sklearn.feature_extraction import text

add_stop_words = ['i', 'hi', 'hello', 'ugh', 'hey', 'just', 'dont', 'thats', 'right', 'people', 'youre', 'got', 'gonna',
                  'time', 'think', 'yeah', 'said', 'lambda', 'device', 'new']
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)
cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(df.Content)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
corpus = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtm.transpose()))
id2word = dict((v, k) for k, v in cv.vocabulary_.items())
```

```python
lda = models.LdaModel(corpus=corpus, num_topics=7, id2word=id2word, passes=50)
lda.print_topics()
```

After some parameter search num_topics = 7 resulted in the optimal topic modelling.
* Topic 0: smart home devices
* Topic 1: device issues and support
* Topic 2: HVAC
* Topic 3: software
* Topic 4: electric panel
* Topic 5: sensors, motion detection
* Topic 6: smart lighting
