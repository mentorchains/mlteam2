```python
import scrapy
from scrapy.crawler import CrawlerProcess
import pandas as pd
import json
import logging
import pathlib

class JsonWriterPipeline(object):

    def open_spider(self, spider):
        self.file = open('data.jl', 'w')

    def close_spider(self, spider):
        self.file.close()

    def process_item(self, item, spider):
        line = json.dumps(dict(item)) + "\n"
        self.file.write(line)
        return item

class IoTSpider(scrapy.Spider):

    name = 'iot'
    start_urls = [
        "https://community.smartthings.com/latest",
    ]

    custom_settings = {
        'LOG_LEVEL': logging.WARNING,
        'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 1}, # Used for pipeline 1
        'FEED_FORMAT':'json',                                 # Used for pipeline 2
        'FEED_URI': 'data.json'                        # Used for pipeline 2
    }
    
    def parse(self, response):
        
        topics = response.css(".raw-topic-link::text").extract()
        tags = response.css(".category-name::text").extract()
        urls = response.css(".raw-topic-link::attr(href)").extract() 
        
        for topic, tag, url in zip(topics, tags, urls):
            yield{
                "topic": topic,
                "tag": tag,
                "url": url
            }
```

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

process.crawl(IoTSpider)
process.start() #
```
