```python
from urllib.request import urlopen as UReq
```

```python
from bs4 import BeautifulSoup as soup
import pandas as pd
dataframe = pd.DataFrame()
```

```python
my_url = "https://community.smartthings.com/latest?no_definitions=true&page=5"
```

```python
uClient = UReq(my_url)
```

```python
page_html = uClient.read()
```

```python
uClient.close()
```

```python
page_soup = soup(page_html, "html.parser")
```

```python
container = page_soup.findAll("tr",{"class","topic-list-item"})
```

```python
'''Here the titles, tags and links are scraped and stored in lists to add them in a dataframe df'''
titles = []
links = []
tags = []
for c in container:
    title = c.a.text
    link = c.a["href"]
    tag = c.findAll("span",{"class","category-name"})
    titles.append(title)
    links.append(link)
    tags.append(tag[0].text)
data = {'titles' : titles , 'links':links , 'tags':tags}
df = pd.DataFrame(data, columns = ['titles','links','tags'])    
```

```python
'''df holds scraped data from individual pages'''
df
```

```python
'''Here dataframe is the final Dataframe that consists of scraped data from latest 5 pages'''
dataframe = dataframe.append(df, ignore_index = True)
```
